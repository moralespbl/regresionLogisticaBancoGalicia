```python
# Convertir en tipo datetime la columna fecha
df_DLR_2014['fecha'] = pd.to_datetime(df_DLR_2014['fecha'], dayfirst=True, errors='coerce')

# Split columna mes entrega:
df_DLR_2014['anio_vto'] = df_DLR_2014.iloc[:, 7].astype(str).str[:4]
df_DLR_2014['mes_vto'] = df_DLR_2014.iloc[:, 7].astype(str).str[4:]

# Covertirlos a int:
df_DLR_2014['anio_vto'] = df_DLR_2014['anio_vto'].astype(int)
df_DLR_2014['mes_vto'] = df_DLR_2014['mes_vto'].astype(int)

# Eliminar columna 7:
df_DLR_2014.drop(columns=df_DLR_2014.columns[7], inplace=True)
```

```python
# Split de la columna TIPO CONTRATO:
df_DLR_2025['mes_vto'] = df_DLR_2025.iloc[:, 1].astype(str).str[3:5].astype(int)
df_DLR_2025['anio_vto'] = df_DLR_2025.iloc[:, 1].astype(str).str[-4:].astype(int)

# Borrar la columna TIPO CONTRATO
df_DLR_2025.drop(columns=df_DLR_2025.columns[1], inplace=True)

# Converir FECHA en timedate
df_DLR_2025['FECHA'] = pd.to_datetime(df_DLR_2025['FECHA'], format='%d-%m-%Y', errors='coerce')

# Reemplazar coma por punto en las columnas y convertir en float
cols = ['PRIMERO', 'MINIMO', 'MAXIMO', 'ULTIMO', 'AJUSTE / PRIMA REF.']


for col in cols:
    df_DLR_2025[col] = df_DLR_2025[col].str.replace(',', '.', regex=False).astype(float)
```

```python
cols = df_DLR_2025.columns.tolist()
# Sacar la última columna
ultima_col = cols.pop(-1)
# Insertar la última columna en la penúltima posición
cols.insert(len(cols) - 1, ultima_col)
# Reordenar columnas en el DataFrame
df_DLR_2025 = df_DLR_2025[cols]
```

```python
# Renombrar Columnas para que coincidan antes de unir:
df_DLR_2025.columns = df_DLR_2014.columns

# Unir los dos DF en uno solo:
df_futuro_dolar = pd.concat([df_DLR_2014, df_DLR_2025], ignore_index=True)
df_futuro_dolar.sort_values(by='fecha', inplace=True)
df_futuro_dolar.reset_index(drop=True, inplace=True)
```
