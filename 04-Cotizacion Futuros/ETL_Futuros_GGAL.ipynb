```python
# Convertir en tipo datetime la columna fecha
df_GGAL_2017['fecha'] = pd.to_datetime(df_GGAL_2017['fecha'], dayfirst=True, errors='coerce')

# Split columna mes entrega:
df_GGAL_2017['anio_vto'] = df_GGAL_2017.iloc[:, 7].astype(str).str[:4]
df_GGAL_2017['mes_vto'] = df_GGAL_2017.iloc[:, 7].astype(str).str[4:]

# Covertirlos a int:
df_GGAL_2017['anio_vto'] = df_GGAL_2017['anio_vto'].astype(int)
df_GGAL_2017['mes_vto'] = df_GGAL_2017['mes_vto'].astype(int)

# Eliminar columna 7:
df_GGAL_2017.drop(columns=df_GGAL_2017.columns[7], inplace=True)
```

```python
# Split de la columna TIPO CONTRATO:
df_GGAL_2025['anio_vto'] = df_GGAL_2025.iloc[:, 1].astype(str).str[-4:].astype(int)
df_GGAL_2025['mes_vto'] = df_GGAL_2025.iloc[:, 1].astype(str).str[4:6].astype(int)

# Borrar la columna TIPO CONTRATO
df_GGAL_2025.drop(columns=df_GGAL_2025.columns[1], inplace=True)

# Converir FECHA en timedate
df_GGAL_2025['FECHA'] = pd.to_datetime(df_GGAL_2025['FECHA'], format='%d-%m-%Y', errors='coerce')

# Reemplazar coma por punto en las columnas y convertir en float
cols = ['PRIMERO', 'MINIMO', 'MAXIMO', 'ULTIMO', 'AJUSTE / PRIMA REF.']


for col in cols:
    df_GGAL_2025[col] = df_GGAL_2025[col].str.replace(',', '.', regex=False).astype(float)
```

```python
# Renombrar Columnas para que coincidan antes de unir:
df_GGAL_2025.columns = df_GGAL_2017.columns

# Unir los dos DF en uno solo:
df_futuro_GGAL = pd.concat([df_GGAL_2017, df_GGAL_2025], ignore_index=True)
df_futuro_GGAL.sort_values(by='fecha', inplace=True)
df_futuro_GGAL.reset_index(drop=True, inplace=True)
```
