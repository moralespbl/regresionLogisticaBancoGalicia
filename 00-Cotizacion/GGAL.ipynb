```python
import pandas as pd 
import yfinance as yf
```

```python
# Crear el objeto del ticker
ticker = yf.Ticker("GGAL.BA")

# Obtener datos hist√≥ricos
GGAL = ticker.history(start="2000-07-26", end="2025-05-16")

# Mostrar las columnas disponibles
print(GGAL.columns)
```

```python
GGAL.head(10)
```

```python
GGAL.to_csv("GGAL.csv", index=True)
```
